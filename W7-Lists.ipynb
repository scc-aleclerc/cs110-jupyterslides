# Python Lists
---

- Python allows variables to be set to *multiple* values
- Variables that hold multiple pieces of data are categorized as *collections*
- The simplest collection is called a **List**

````python
newlist = ['a', 'b', 'c']
````

The above list is stored in the variable name `newlist` and it contains three values, the strings `'a'`, `'b'`, and `'c'`

## Lists Definition
- Made up of *items* / *elements* (interchangeable)
	- Each element holds one value
- Sequential
	- List values are held in a *defined order*
- Mutable
	- The elements can be changed

## Syntax
```python
numbers = [1, 3, 8912, 028]
```
- Must have values in between square brackets: `[` `]`
- Values are separated by a comma: `,`

# Indexes
- Lists are *indexed*
- Each item is assigned a number which represents its order in the list
- To access an item in the list, we have to retrieve it using it's number within the list:
```python
listname[0]
```
- The `listname` is whatever the lists variable name was defined as
- The number in the square brackets is the index, aka, the position of the item in the list
- **Indexes start counting from 0!**

````python
fruits = ['apples', 'bananas', 'cantalope']
print(fruits[0])
````

Counter intuitively, if we have a `list` with 3 items, using `list[3]` will return an `IndexError`

````python
fruits = ['apples', 'bananas', 'cantalope']
fruits[3]
````

- Why do we think this happens?

- Because lists start counting from `0`, the biggest index we can use is:
	- length - 1
- Where length is the total number of items in the list

# `len()`
- Most programming languages that have collections / lists, have a function that gives the *length*
- In python, the function is `len()`
	- The variable name of the list must be put into the paratheses `( )` 

````python
fruits = ['apples', 'bananas', 'cantalope']
print(len(fruits))
````

# Changing List Items
- So far we've *read* values stored in a list
- We can also add, remove, and change items in a list

- Lists are basically just variables grouped together and referenced with an index
- We can think of the variables as all living in the same apartment complex, and their index being the unit number

## Update
- We can treat the items in a list just like any other variable by referring to the list name and the index of the item we want to change

````python
stationary = ["pen", "pencil", "eraser", "highlighter"]
stationary[3] = "marker"
print(stationary)
````

## Add Items
- We can add items to a list, usually onto the end or the beginning
    - *Appending* to the end
    - *Prepending* to the beginning

````python
groceries = ["almonds", "bread", "carrot"]
groceries.append("donuts")
print(groceries)
````

## Removing Items
- We can either remove items by using `del` on a specific index:

````python
candy = ["chocolate", "lollipop","gumdrops"]
del candy[1]
print(candy)
````

- We can remove a value from the list

````python
jewelry = ["necklace", "bracelet", "ring", "earring"]
jewelry.remove("necklace")
````

- If we use `remove()` on a list that has *duplicate* values in it, it will remove the first value it comes across.

````python
jewelry = ["necklace", "bracelet", "ring", "bracelet", "earring"]
jewelry.remove("bracelet")
````

## Checking Membership
- We can check if a specific value is stored in a list using `in`

````python
amphibians = ["axolotl","newt","frog","salamander"]
if("frog" in amphibians):
	print("is an amphibian")
````

We can also check if a value is `not in`

````python
amphibians = ["axolotl","newt","frog","salamander"]
if ("turtle" not in amphibians):
	print("not an amphibian")
````

- `in` and `not in` return *Boolean* 
	- that's why we can use them with `if`
- when using `in` / `not in` be very careful of spelling!
	- a list containing the string "frogs" or "Frog" is different than the word "frog"
    - (there are more complex ways of checking against strings, such as regular expressions, but we are not going to worry about that right now)

## Heterogeneous Lists
- Lists can hold items of *differing types* in Python
- This is unique to python
	- most programming languages require you to define the type of data your list will hold, and won't let you deviate

````python
different = [1.2, "hello", True]
print( type( different[0] ) )
print( type( different[1] ) )
print( type( different[2] ) )
````
