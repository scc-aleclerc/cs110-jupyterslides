# If Statements
- In programming, we often want to have different behaviors depending on the data we have
- This is referred to as **conditional**
	- If a condition is met, do this behavior
- Or **branching**
	- The flow of the program *branches* depending on what happens

## If Syntax
 ```
 if (<condition>):
	 <do...>
 ```
- Python uses *indents* to understand relationships between lines of code
- An `if` must be followed by an indented line
    - Anything indented underneath the if statement is called being "*within* the `if` statement"
- `IndentationError` will be thrown if Python can't understand the indentation

````python
from random import randint

secret = randint(0,100)

print(secret)

if(secret > 50):
	print("greater")

if(secret < 50):
	print("lesser")
````

Each time the above code is run, it choose a number between 0 and 100 and then prints whether that number is greater than or less than 50

- What happens if the variable `secret` is exactly 50?
- (you could edit the code above with: `secret = 50`)

    - If `secret` is exactly 50, the code does nothing. 
    - It fails both `if` statements, so it does no behavior.

## INCLUSIVE checks
- When checking if something is *less than* or *greater than*, we have to be precise
- How does a computer interpret: "Give me all of the numbers from 1 to 50"?
	+ Including 50? (1...50)
	+ Excluding 50? (1...49)

## Example
If we want to program this retail promotion:

\$1.00 off any purchase over \\$5.00!

\$5.00 off any purchase over \\$20.00!

\$20.00 off any purchase over \\$50.00!

````python
spent = 9.98
discount = 0.00

if (spent > 5.00):
	discount = 1.00
	
if (spent > 20.00):
	discount = 5.00

if (spent > 50.00):
	discount = 20.00

print(discount)
````

- The above code appears to work
- BUT, there's an issue if the `spent` variable is set to the exact cutoff...

- If the `spent` variable's value is set to an exact cuttoff
- For instance, if `spent == 20`, discount will be set to `1.00` instead of `5.00`!

````python
spent = 20.0
discount = 0.00

if (spent > 5.00):
	discount = 1.00
	
if (spent > 20.00):
	discount = 5.00

if (spent > 50.00):
	discount = 20.00

print(discount)
````

- This is because if we think about it, the second if statement is evaulating:
- Is`20.00 > 20.00`?
- The answer is no! `20.0` is not greater, but it is greater than `5.00`, so it keeps the discount set to `1.00`, which it was set to in the previous `if` statement.

We can visualize the range as it is, like this:
![](img/w4-conditionals-discount1.png)
- The longer notches represent our cutoffs (5.00, 20.00, and 50.00)

## ...or equal to
- If we combine the two symbols...
    - `<=` : Less than or equal to
    - `>=` : Greater than or equal to
- We can check *up to and including* that number

\* The `>` or `<` is always on the *left* side of the `=`

````python
spent = 20.00
discount = 0.00

if (spent >= 5.00):
	discount = 1.00
	
if (spent >= 20.00):
	discount = 5.00

if (spent >= 50.00):
	discount = 20.00

print(discount)
````

This is the range after using `>=`...
![](img/w4-conditionals-discount2.png)

We could also cover the desired range using JUST `>`

````python
spent = 9.98
discount = 0.00

if (spent > 4.99):
	discount = 1.00
	
if (spent > 19.99):
	discount = 5.00

if (spent > 49.99):
	discount = 20.00

print(discount)
````

(This isn't my preferred way to do this for price ranges, because we may do calculations on prices that result in a price that's something like 49.999999999, which could still pass the above `> 49.99` check.)

## Why use `if`?
- If statements check if the condition it's given is *true*
	- For instance: is this value greater than a threshold?
	- Are two values equal?
- This is useful for evaluating *values we can't control*
	- For example: User input, data from outside the program

## What if we want to check in between a range?
- For instance, all numbers between, and including, 10 to 20?
- Above, we looked at strategically placed `if` statements.
- In reality, for values greater than or equal to `50` in the code above, the `discount` variable gets overwritten multiple times
    1. `spent = 72.00`
    2. `spent >= 5.00`, `discount = 1.00`
    3. ALSO `spent >= 20.00`, overwrite: `discount = 5.00`
    4. ALSO `spent >= 50.00`, overwrite: `discount = 20.00`
The `discount` has changed values 3 times, but we don't see it until the final value is `print`'d at the end.

## Is it bad?
- This method still works for our purposes, because we are just trying to print a final value.
- We don't see the changing value of discount, and it doesn't matter until we do something with the final value
- HOWEVER, this code will be prone to error if it's ever touched because it's accuracy depends on the *ordering* of the if statements

If we print in each if statement, we can more clearly see the issue...

````python
spent = 79.98
discount = 0.00

if (spent >= 5.00):
	discount = 1.00
	print(discount)
	
if (spent >= 20.00):
	discount = 5.00
	print(discount)

if (spent >= 50.00):
	discount = 20.00
	print(discount)
````

# Combining Conditions
- Luckily, we can combine conditions using *logical operators*

## `and` Operator
- Both conditions MUST be true 
	- (result in a true value once computed)
- If EITHER condition is false, the statement returns false


````python
a = 5
b = 5

if (a == 5 and b == 5):
	print("both 5")
````

(the keyword `and` must be in all lowercase to work in Python)

- If even one of these conditions is not met, we fail the `if` statement and nothing happens

````python
a = 1
b = 5

if (a == 5 and b == 5):
	print("both 5")
````

We can also use logic to check the same variable (not just compare two different variables)

````python
a = 10
if (a > 0 and a < 20):
	print("good")
````

## `or` Operator
- At least one condition on either side must be `True`
- Both can be true, and the then the statement will be considered `True`
- When we use the word "or" in English, we usually mean "one or the other"
- Allowing for both is called an *inclusive* `or`

````python
x = 1
y = 5

if ((x == 1) or (y == 1))
	print("a one")
````

Also works if both are 1...

````python
x = 1
y = 1

if ((x == 1) or (y == 1))
	print("at least one")
````

## `xor` Operator
- *E**x**clusive* `or` operator
- ONLY one of the conditions on either side can be True
	- if both are `True`, the operator counts the whole statement as `False`
- In python, represented by a `^` (carrot)
    - It doesn't work on all data types... it's a little tricky in Python, and not always used. 

````python
a = 1
b = 2
print((a == 1) ^ (b == 1))
````

````python
a = 1
b = 1 
print((a == 1) ^ (b == 1))
````

## Remember: We ❤️ Parentheses!
- Parentheses are extra important when we're using logical operators
- We need to be clear to best anticipate the behavior
- Code needs to be able to be read by other humans
	- Can't assume that everyone always correctly understands how the order of operations resolves

# Else (and Elif)
---

## Limitations of `if`
- With just `if` statements, we either do or don't do a behavior
- We have to make sure the condition could pass *another* `if` for alternate behavior
- and, sometimes a condition may pass *multiple* `if`s...

````python
spent = 79.98
discount = 0.00

if (spent >= 5.00):
	discount = 1.00
	print(discount)
	
if (spent >= 20.00):
	discount = 5.00
	print(discount)

if (spent >= 50.00):
	discount = 20.00
	print(discount)
````

## Solutions
- make the if statement more strict using `and` / `or`
	- `if( (spent >= 5.00) and (spent < 20) )`
	- con: need good understanding of ranges to prevent errors
		- may accidentally miss more *edge* cases
- consider whether you care about *end* result vs correctness through the program
	- originally the code printed after all the if was done, so we didn't see when discount was incorrect for large values

## Else
- There is an optional second part of an `if` statement
- `else` allows us to define behavior for *all conditions that don't meet the if*

````python
from random import randint

threshold = randint(0,100)
print(threshold)
if(threshold > 60):
    print("above")
else:
    print("below")
````

## Guidelines
- a lone `if` will compute following indented lines ONLY if the condition passes
- `else` will follow the lines indented under `else` in all cases where `if` fails
- `else` statement ==needs to directly follow the `if`== , they work in tandem
- `else` acts as the default behavior when `if` fails. 
	- without `else`, nothing happens

## Elif (Else if)
- `elif` allows us to try *another* check when (and only when) a first `if` statement fails
- the code inside the `elif` only executes when it's condition is true
- we can have as many `elif` statements as we want, including 0

````python
number = 9

if((number > 0) and (number < 5)):
	print("low")
elif((number >= 5) and (number <= 15)):
	print("middle")
elif((number > 15) and (number <= 20)):
	print("high")
````

## Overlapping
- when an `if` and `elif` have overlapping `True` conditions, the one that comes first will compute

````python
number = 8

if((number > 0) and (number <= 10)):
	print("aaa")
elif((number > 5) and (number < 20)):
	print("bbb")
````

![](img/w4-conditionals-overlap1.png)

- The range in the `if` takes priority because it comes before 
	- code is read / computed from top to bottom
- The `elif` *only* checks the condition when the `if` fails
	- but, the way it is written it will still check the part of the range already covered by the if

## An easier range
When we want the condition to trigger *both* statements, we should use two `if` statements, not an `elif`

````python
number = 8

if((number > 0) and (number <= 10)):
	print("aaa")
if((number > 5) and (number < 20)):
	print("bbb")
````

## Leveraging Overlap
This overlapping quirk is actually in our favor for shorter / cleaner code when we understand it...

````python
number = 7

if((number > 0) and (number <= 5)):
	print("no")
elif(number =< 10):
	print("ok")
elif(number =< 15):
	print("yep")
````

![](img/w4-conditionals-overlap2.png)
Because the statements are always checked *top to bottom* first we can 'stack' conditions if we understand the order

## Else as Catch-all
- Often, we want to use `if`/`elif`/`else` statements to make sure that our data is *valid*
- Usually we think of `if` and `elif` as getting the *positive* cases (ones we are expecting)
- We can use `else` to catch everything else

````python
limit = 6
var = 0

if(limit <= 5):
	print("first")
elif(limit <= 10):
	print("second")
else:
	print("other")
````

## Case statements
- Many programming languages have *case statements* 
	- aka switch statements
- They basically work like an `elif` that checks for a lot of `==`

### Switch example in Java
```java
int day = 4;
switch (day) {
  case 1:
    System.out.println("Monday");
    break;
  case 2:
    System.out.println("Tuesday");
    break;
  case 3:
    System.out.println("Wednesday");
    break;
  case 4:
    System.out.println("Thursday");
    break;
  case 5:
    System.out.println("Friday");
    break;
  case 6:
    System.out.println("Saturday");
    break;
  case 7:
    System.out.println("Sunday");
    break;
}
// Outputs "Thursday" (day 4)
```

# Cases in Python

We can write the previous Java code in python using `elif`

````python
day = 4;
if(day == 1):
	print("Monday")

elif(day == 2):
	print("Tuesday")

elif(day == 3):
	print("Wednesday")

elif(day == 4):
	print("Thursday")
	
elif(day == 5):
	print("Friday")

elif(day == 6):
	print("Saturday")

elif(day == 7):
	print("Sunday")
````

We can also use the brand-new (as of Python version 3.10) `match` cases:

````python
day = 4
match day:
	case 1:
		print("Monday")
	case 2: 
		print("Tuesday")
	case 3: 
		print("Wednesday")
	case 4: 
		print("Thursday")
	case 5: 
		print("Friday")
	case 6: 
		print("Saturday")
	case 7: 
		print("Sunday")
````

## Catch all: `case _:`
- Just like our else, we can use this for all values that don't fit a specific case
- Any time the variable being evaluated doesn't match a case, it will trigger the `_` case

````python
day = 12
match day:
	case 1:
		print("Monday")
	case 2: 
		print("Tuesday")
	case 3: 
		print("Wednesday")
	case 4: 
		print("Thursday")
	case 5: 
		print("Friday")
	case 6: 
		print("Saturday")
	case 7: 
		print("Sunday")
	case _:
		print("Day does not fit week day numbers")
````

## Elif by another name
- Functionally, using a bunch of `elif` statements is about the same as using a `case`
- Technically, if you have a lot of conditions, a `case` statement is more efficient
- However, if you actually have enough statements (thousands, or more...) you probably have other bigger problems to solve
