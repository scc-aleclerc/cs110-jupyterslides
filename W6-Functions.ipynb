# Functions
---

## Functions in Python
- Functions (sometimes referred to as methods) are blocks of statements that we group together with a *name*
- Similar to a variable, but instead of giving a name to values, we group together *behaviors* under a name

## Anatomy of a Function
```python
def name( parameters ):
	statements
```
- We need to include `def` as the keyword to Python that we are **def**ing a new function
- We give the function a *name*, similarly to how we name variables
- Functions must be followed by parentheses (`( )`)
- We can give functions *parameters*, which will be treated like local variables inside the function
- We must follow the parentheses with a colon (`:`)
- Any and all code we want to execute in the function happen in *indented* code statements

## Components
- The *header* defines the name of the function as well as any (optional) parameters
	- We have to use the `def` keyword to signal to python that we want to **define** a new function
	- Must end with a colon `:`
- The *body* defines the actual actions (code) that performs the functions 

## Indentation
- Python understands which pieces of code belong to the function using *indented* lines underneath the header line
- What else have we encountered that used indentation like this?
- `if` statements similarly define groups like this

## *Using* functions
- There are two steps to using a function:
	- Defining
	- Calling

````python
def manythanks():
	print("thanks")
	print("thanks")
	print("thanks")
	print("thanks")
	print("thanks")
````

## Wait!
- Nothing happens when we run the above code, even though there are no errors
- We have to *call* the function in order for anything to happen
- Defining just instructs python *what to do later*, we have to call it for it to run

````python
def manythanks():
	print("thanks")
	print("thanks")
	print("thanks")
	print("thanks")
	print("thanks")
	
manythanks()
````

## Call and Response
- The function definition is like a *sing-a-long*
- We're instructing python on how the chorus goes
	- even if it's not time to use it quite yet
- The band can play the verses (other code statements), but call back to the *chorus* (the function) at the right time

## Built-in Functions 
- We've already used some functions, but we didn't define them
- `print()` is a function, it was written by the authors of the programming language rather than us

## Use case: Repeated behavior
- One benefit of functions is that we can *group* together behaviors, and easily repeat them

````python
def greeting():
	print("hello!")
	print("hey!")
	print("hi!")

greeting()
greeting()
greeting()
````

Alternatively, we could just repeated the print statements 
(copy/paste them) 
it would work the same

````python
print("hello!")
print("hey!")
print("hi!")
print("hello!")
print("hey!")
print("hi!")
print("hello!")
print("hey!")
print("hi!")
````

This works, but there are some critiques

- creates LOTS of lines of code
- harder to change later
	- if we want to repeat twice instead of 2 times, we have to carefully remove 3 lines of code instead of 1 function call
- if there is ever a change / mistake, we have to change *all* copies of the repeated code, in a function we just have to change the function statement

## Parameters
- Parameters allow us to define variables that the function will use
	- Parameters are optional

````python
def greeting(name):
	print("hello", name,"!")
	print("hey", name, "!")
	print("hi", name,"!")
	
greeting("Sprout")
````

## Parameters as Local Variables:
- We can also use *variables* to pass into functions
- Remember, python just replaces anywhere it sees the variable with it's value when it actually computes

````python
def greeting(name):
	print("hello", name,"!")
	print("hey", name, "!")
	print("hi", name,"!")
	

user = "Sprout"
greeting(user)
````

## Parameters Create Flexibility
- Parameters act as a placeholder for behavior that can be replaced later
- We can repeat the same *behavior* with different *values*

````python
def greeting(name):
	print("hello", name,"!")
	print("hey", name, "!")
	print("hi", name,"!")

x = "Sprout"
y = "Genny"

greeting(x)
greeting(y)
````

## Important: Positioning

- Technically, you can define functions wherever you want in a program
- However, usually functions are defined *at the top* of a program
- Python reads code top to bottom, so functions need to be *defined* before they are called
- Think about the sing-a-long example: would it make sense for the band to play the whole song, and then explain the sing-a-long portion after the fact?

# Returning Output
---

- Functions can *optionally* have parameters
	- Data to go into the function
- Functions can also *optionally* have a return value
	- Data that comes out of a function

## Syntax
```python
def name( parameter(s) ):
	...
	return value
```
- If we want the function to give a value back to the part of the program it was called from, qe use the `return` keyword
- We follow the `return` with any values we want to give back 

````python
def iseven(number):
	if(number % 2 == 0):
		output = True
	else:
		output = False
		
	return output
````

## Calling the Function

````python
print(iseven(8))
````

## Return Ends Functions
- A return statement ends the computation of a function
- This is obvious when we put the return statement at the bottom
- However, it is possible to return elsewhere in the function

## `return` Pattern: Return Variable
- One pattern we can follow is holding the return value in a variable, and then returning the variable at the end of the function

````python
def grade(points):
	if(points >= 60):
		if(points >= 90):
			output = 'A'
		elif(points >= 80):
			output = 'B'
		elif(points >= 70):
			output = 'C'
		else:
			output = 'D'
	else:
		output = 'F'
	return output
````

## `return` Pattern: Return Value
- In the following piece of code, we can just return the value directly from the relevant condition
- Remember: **using `return` ends the function immediately**, so any code after the return will not be run.

````python
def grade(points):
	if(points >= 60):
		if(points >= 90):
			return 'A'
		elif(points >= 80):
			return 'B'
		elif(points >= 70):
			return 'C'
		else:
			return 'D'
	else:
		return 'F'
````

## Scope
- Parameters only exist within the function they belong
- This is called *scope*
- We can't use parameters outside of their function, it is out of scope

````python
x = 0

def examplescope(x):
	print(x)

pocket(12)
print(x)
````

## Local Scope
- Variables defined inside of functions are not accessible outside of functions
- They are created when the function is called, and end when the function completes (*local* to the function)
- This is referred to as the *lifetime*
- (This applies to other, more sophisticated components as well)

## Scope Error
- The following code results in an error
- The variable `y` cannot be used outside of the function because it was *defined* and *initialized* inside the function.
- The code in the function does not run until the name of the function is called to use

````python
def pocket():
	y = 6
	y = 6 * 2
	
print(y)
````

## Global Scope
- Technically, within a function we can reference variables that are defined outside of said function
- These are referred to as **global** variables or **global** scope


````python
def badsquare(x):
	output = x ** power
	return(output)

power = 2
print(badsquare(2))
````

## Avoid using Globals
- This is technically possible in python
- However, **using global variables is best avoided**
- Global variables create *security weak points* 
- They make code difficult to expand out
- For our learning purposes, you can use them to solve problems, but should try to design alternatives and not be too reliant on them

## Alternatives to Globals
- Instead, it is better to use *parameters* and *return* to pass things in and out of functions
- This makes functions more versatile (they can be used by other programs)
- It also helps protect against security flaws

````python
def oksquare(x, y):
	output = x ** y
	return output

z = oksquare(2,4)
print(z)
````

## Local vs Global Priority
- First, python checks if there are any *local* variables it can used within the function
- If it fails to find any, it then checks global variables
- This means that local variables will always have priority, even if you have two variables of the same name inside and outside a function

````python
x = 5

def multitude(y):
	x = 2
	output = y * x
	return output
	
print(multitude(2))
print(x)
````

## Scope Order of Operations
- Python *always* checks local variables **first**
- In the previous example, it considered the `x` inside the function to be the creation of a *local* x variable
- The term for this is *shadows*
- The local `x` *shadows* the global `x`
	- It makes the global `x` inaccessible

## None as a Value

- This is actually distinct from zero

- *nothing* is the absence of a value
	- also referred to as None, Nan, Null, Nil, $\varnothing$, etc

## Functions Calling Each Other
- Functions can call other functions to complete their computation
- We've already seen this; `print()` is a function, we just didn't write it, but we used it in functions
- We can have functions we write call other functions we wrote

````python
def iseven(x):
	if(x % 2 == 0):
		return True
	else:
		return False
		
def evendivide(y):
	if(iseven(y)):
		return y / 2
	else:
		return None
````
