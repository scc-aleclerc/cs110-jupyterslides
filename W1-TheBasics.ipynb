## What is a *program*?
- A recipe that explains how to complete a task
- Ordered set of instructions
- Must have a "terminal" end point
  - This means there must be a state that defines the task as done

## What is *programming*?
- Writing recipes in a language that computer can understand
- Several points of *translation*:
	- from a *high level* programming language humans can read
	- to a *low level* language the computer can read
		- all computers speak a native language of **binary**

![A figure with three bubbles, an arrow pointing down from the top to the next, and the middle to the bottom. Too bubble labeled high level, programming language. Middle bubble labeled low level, assembly language. Bottom bubble labeled machine level, binary code.](img/W1-languages.png)

## Abstraction
- Learning to code means writing programs that can be applied to solve problems
- Ideally, we create programs that can solve *different* but *similarly formatted* problems
- *Abstraction* is the process of breaking a problem into small, modular, solvable pieces
- Through programming, we learn how to use **abstraction** to break a *complex* problem into manageable pieces that can be addressed with code

## Scope of this Course
- In this course, we will focus on high level programming
- This is usually what people mean when they say "coding", "programming", or "developing"
- It's useful to *understand* lower-level programming, but it is used less often

## Talking to Computers
- In programming we are trying to communicate our problem-solving recipe to the computer
- Often, we have mistranslations that lead to errors
- We eventually learn how to write code in such away that minimizes errors and loopholes
