## Variables in Other Languages
- In other programming languages, we declare a variable like this:
```
int coolNumber = 72;
```


- Notice the **`int`** before the variable name

## Specific Data Types
- In these other languages, we have to tell the program what exact data we are going to store in that variable.
- If we think about our old box analogy, these programming languages have specifically shaped boxes for each type of thing we want to store

![](img/w2-datatypes-different.png)

## Defining Variable Types in Other Languages
- In these programming language, the computer needs a command that tells it *what kind of box* to use for the data
- "Go get a round box to store this hat in, and label it `coolhat`"

## Defining Variable Types in Other Languages
- In python, we don't need to tell the computer what kind of data we are storing
- Python *guesses* the data we are using, based on context clues
- For instance: 
```
var = 3
```
- Python *assumes* that this data is an integer, because the variable is holding data that seems like an integer

## Flexible Data Types in Python
- In other programming languages, once we set a variable to a certain type, that variable can only store data of that type
	- For instance, once we say a variable is an *integer* variable, we cannot store *strings* (text) in it.
- Because python's variables (boxes) are *non-specific*, (technical term: dynamic, polymorphic), a single variable can change types / hold different data whenever necessary

## Bad Java
```
int var = 12;
var = "hello world";
```
- In the Java programming language, this results in an error.
- We promised Java we would use the variable `var` to store *only* integers.


## Fine Python

````python
var = 12
var = "hello world"
````

- The above code works fine in python
- Python doesn't care about the type of data that used to be stored in a variable, it just adapts

## Data Types in Python
- Does this mean that python variables **don't** have types?

- No! Python variables *do* have types
- Python just adapts the data type based on the context of what is stored in a variable

## Finding Types
- We can use the following function to display the type of the data stored in a variable:

````python
var = 12
print( type(var) )
````

- (we can ignore the word `class` for right now)
- Here `int` refers to *int*eger

## Types of Numbers

````python
a = 24
print( type(a) )

a = a / 4

print(a)
print( type(a) )
````

## Decimals = Floats
- In programming, a decimal point can be represented in two ways. Most commonly, it is a **floating** decimal point.
- In python, the *division* operator outputs a *float* type
- A lot of division ends in a *decimal* number, so the operation outputs a number with a decimal (even if it is round)
- Round numbers get printed with a `.0` as an indicator of what *data type* it is.

## Another Kind of Division
- Usually, using the `/` for division is fine
- Python provides *another* way to do division

````python
print(15 // 2)
````

- This method gives an *int* result instead of a *float*
- It drops any numbers beyond the decimal point
    - It doesn't *round*, it just omits any numbers after the decimal point
    - It is less precise because of this
