## Learning to Program
- Learning to program is a very *iterative* process.
    - (Iterative: involves multiple repetitions)
- At first, you are going to just practice running code without fully understanding how or why it works
- We will build knowledge over time and fill in the gaps as your understanding becomes richer

## Learning to Program
As you are learning to code, keep in mind:
- Your code solutions will look different than mine
- Your code solutions will look different than your peers
- There are different ways to arrive at the same answer
Often in programming, there are different but equally valid solutions to the same result.
