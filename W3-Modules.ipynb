## Importing Functionality
- All programming languages have **built-in** functionality and additional functionality, usually called **libraries**, **packages**, or **modules**.
- Built-in functionality is what comes packaged with the basic programming language
- Often, you may be restricted by the basic functionality
- From here, you have two options:
	- Write the functionality yourself
	- Use code other people have created to help you solve your problem

## When to Import
- If you want to solve a problem with pre-established or standard practices, likely someone (or multiple people) have well-written and mature library for you to use
- For instance: statistics, interfacing other technology, machine learning, etc 

## When to DIY
- Later we will learn how write our own functions (like the `print()` or `type()` function)
- Sometimes, you are programming in a very specific environment with very specific data / processes / needs etc
	- For instance: at a job, where data (ID numbers, vocabulary, processes) are distinct and predetermined

## Where are the libraries?
- Python comes with *the basic functionality* and installs (but does not always load in) many commonly used modules 
- You can download and use add-on libraries that are *installed with python* but not always loaded into your program
	- This is for space and efficiency
- Officially, python libraries are called *modules*
	- It is helpful to know the official word when reading official documents or errors, but note that many people will refer to modules as libraries/packages interchangeably.

## Imported Names
- If we just use `import` , then we have to lead the function we want to use with the name of the module followed by a period (`.`).

```python
import math
print( math.sqrt(16) )
```

- Now it works!

## Importing Under a New Name
- We can change the name of the module by doing this:

```python
import math as m
print( m.sqrt(49) )
```

- If you are using two libraries that may have functions of the same name (like the `math` and `pandas` modules) then it is helpful to keep the names, but you can shorten them to cut down on how much you have to type

We can select specific functions we want to import from the math module by listing them like so:

```python
from math import pow
print( pow(2,8) )
```

- Notice, when we import like this, we don't have to prepend the function name with the module name

- We can choose multiple functions out of a module if we use a comma in between the name of the functions:

```python
from math import pow, sqrt
print( pow(12,4) )

print( sqrt(20736) )
```

- We can just load in *all* the functions of a module if we do this:

```python
from math import *
print( log(16,2) )
```

## `*` in Computer Science
- Often, in different parts of Computer Science, the \* (referred to as a *star*) means: all inclusive, from 0 to all possible quantities
- In this context, it means "load in everything from this module"

## How much to load in?
- Whether you choose to load in all of the stuff from a module or just select a few, will depend on:
	- Context
	- Conflicts with other libraries
	- How complex your solution is
	- Readability
	- Personal preference

## Should you load in ALL the modules?
- After you know what you're doing in your code, best practice is to only import the functions you need from a module
- HOWEVER... when you are learning, you may not know exactly what you need to import.
    - For instance, if you are trying to code a math equation, you may not know which math functions you need at first. You can start by using `from math import *` until you know which ones are useful, to keep you from having to change what names are included in the import.

## What's in a module?
- When you use `import`, the module can have *functions* AND *variables*
	- (and other more advanced stuff we aren't talking about right now)

```python
from math import *

print( gcd(60,72) )
print( pi )
```

- In the above example, `gcd()` is a function
- `pi` is a variable (but it's defined within the `math` module, and then we imported it into our code)
