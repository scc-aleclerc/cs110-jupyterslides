## Turtle
- When using programs to draw images, we usually give instructions for the cursor to move
- In the Python Turtle module, we view this as a turtle, which we tell how to move across a canvas

## \* Turtle in Colab

- Luckily, someone made a Colab-specific version of turtle
- We will use the ColabTurtlePlus module
- The set-up is slightly different than the normal version of turtle, so be aware of that when looking up other resources

## Moving the Turtle
- We treat the turtle as our *cursor* across the canvas
- We move the turtle relative to the center of the canvas 

## Basic Movements
- `forward(<units>)` moves the turtle *forward* in the direction it's currently facing by a number of units (pixels)
- `left(<degrees>)` rotate the turtle to the *left* (counter clockwise) by a number of degrees
- `right(<degrees>)` rotate the turtle to the right (clockwise) by a number of degrees

## Draw a Square Using a Loop
- When we have code that has **repeating** lines, we should use a **loop**
- Anytime we have to repeat a code statement more than twice, we should change the code to include either a **function** or a **loop**
    - In this case, a loop makes sense

```python
# We don't have to import if we have already run a code cell that uses the import
# from ColabTurtlePlus.Turtle import *

clearscreen()
initializeTurtle()

for i in range(4):
	forward(100)
	left(90)
```

## Moving without Marking
- We can think of the turtle like a retractable pen - we can tell it to have the pen up (`penup()`) or down (`pendown()`)
- We want to pull the pen up to move the turtle in case we want to have a gap inbetween lines
- If we keep the pen down, we have to draw like an etch-a-sketch

```python
initializeTurtle()

penup()

forward(100)
left(90)
forward(100)

pendown()

for i in range(4):
    forward(100)
    left(90)

penup()
right(180)
forward(100)

pendown()
for i in range(4):
    forward(100)
    right(90)
```

## Other Useful Features:
- change the background color
    - `bgcolor()`
- built in shapes, like circles, rectangles, and polygons
    - `circle(radius)`, `polygon(sides, length)`
- fill shapes with color
    - `fillcolor(color)`
- create multiple turtles


```python
clearscreen()
setup(300,300)
bgcolor('black')

# creates a new turtle
t1 = Turtle()

t1.color("red", "yellow")
t1.shape("turtle")

t1.pensize(2)
t1.speed(7)

t1.begin_fill()
for i in range(4):
  t1.forward(100)
  t1.left(90)
t1.end_fill()

t2 = Turtle()
t2.shape("turtle")
t2.penup()

t2.left(180)
t2.forward(50)

t2.color("hotpink", "orange")
t2.pensize(2)

t2.begin_fill()
t2.circle(50)
t2.end_fill()
```
