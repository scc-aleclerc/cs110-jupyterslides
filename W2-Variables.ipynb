# Assigning a Value
---
- A single `=` (equals sign) is used differently in programming than it is in other contexts.
- We are used to thinking of a `=` as a check for *equality* or a *statement of fact*
	- e.g. 2 + 2 = 4
- However, in programming we use the `=` to declare that a variable will store a specific value
- This is called *assigning a variable*
- With `coolbox = "cool text"` we *assign* the text value `"cool text"` to the variable name `coolbox`
- Assignment **always** takes the *value* on the *right hand side* and assigns it the *name* on the *left hand side*

#### CODE BLOCK 2

```python
coolbox = "cool text"
```

# Variable Behavior
---
- Variables are sometimes counter intuitive when we first start learning to think like a computer

## Changing Values
- We can change the *value* of a variable at any point in the program
- Usually, we have to use *assignment* (`=`) to change the variable
- Some functions can change the value of a variable without reassigning, but we'll worry about that later

#### CODE BLOCK 9

```python
var = 9
print(var)

var = 12
print(var)

var = 15
print(var)
```

## Arithmetic and Changing Variables
- Using arithmetic on a variable *doesn't change the value* unless we also reassign the variable to the new value

#### CODE BLOCK 10

```python
var = 2
print(9 * var)

print(var)
```

## Change a Variable Using It's Current Value
- A variable can be assigned a new value by using it's current value

#### CODE BLOCK 11

```python
var = 24
var = var / 4
print(var)
```

## Order of Operations
- The computer does all the work necessary on the *right hand side* of the `=` first
- It stores the final value in the variable name after doing any work using the old value on the right hand side first
- Using a variable to transform another variable doesn't change the original variable

#### CODE BLOCK 12

```python
a = 7
b = 2

a = a * b
print(a)
print(b)
```
